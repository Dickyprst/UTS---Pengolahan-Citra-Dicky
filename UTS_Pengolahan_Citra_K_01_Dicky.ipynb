Operasi Flipping

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('dicky-img.png')
img_flip_ud = cv2.flip(img, 0) #flipping secara vertikal, pencerminan pada sumbu x
img_flip_lr = cv2.flip(img, 1)  #flipping secara horizontal, pencerminan pada sumbu y
img1 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #harus konversi dulu dari BGR to RGB kalau mau make plot
img2 = cv2.cvtColor(img_flip_ud, cv2.COLOR_BGR2RGB) #harus konversi dulu dari BGR to RGB kalau mau make plot
img3 = cv2.cvtColor(img_flip_lr, cv2.COLOR_BGR2RGB) #harus konversi dulu dari BGR to RGB kalau mau make plot
plt.subplot(131), plt.imshow(img1), plt.title("Original") #plot memiliki 1 baris dan 2 kolom. citra akan diletakkan pada plot pertama
plt.xticks([]), plt.yticks([])
plt.subplot(132), plt.imshow(img2), plt.title("Hasil flip vertical") #plot memiliki 1 baris dan 2 kolom. citra akan diletakkan pada plot kedua
plt.xticks([]), plt.yticks([])
plt.subplot(133), plt.imshow(img3), plt.title("Hasil flip horizontal") #plot memiliki 1 baris dan 2 kolom. citra akan diletakkan pada plot kedua
plt.xticks([]), plt.yticks([])
```
